```python
##Exercise 1: Perform List Manipulation.
my_list = [10, 20, 30, 40, 50]
# 1) Change Element: Change the second element of a list to 200 and print the updated list.
my_list[1]=200
print(my_list)
# 2) Append Element: Add 600 o the end of a list and print the new list.
my_list.append(600)
print(my_list)
# 3) Insert Element: Insert 300 at the third position (index 2) of a list and print the result.
my_list.insert(2,300)
print(my_list)
# 4) Remove Element (by value): Remove 600 from the list and print the list.
my_list.remove(600)
print(my_list)
# 5) Remove Element (by index): Remove the element at index 0 from the list print the list
my_list.pop(0)
print(my_list)
```

```python
##Exercise 5: Find Maximum and Minimum.
#Find and print the largest and smallest number in a list [8, 2, 15, 1, 9].
import math
MAX=-math.inf
MIN=math.inf
num=[8,2,15,1,9]
for i in range(len(num)):
    if MAX<num[i]:
        MAX=num[i]
    if MIN>num[i]:
        MIN=num[i]
print("max value: ",MAX)
print("min value: ",MIN)
```

```python
##Exercise 13:  find the min element from the row and max element form the col in a 2d list.
import math 
arr=[[0.1,2,3],
     [4,0.5,6],
     [7,8,0.9]]
minOfrow=[]
maxOfcol=[]
# finding the minOfrow
for x in range(len(arr)):
    mini=math.inf
    for y in range(len(arr[x])):
        if mini>arr[x][y]:
            mini=arr[x][y]
    minOfrow.append(mini)
print(minOfrow)
# finding the maxOfcol
for x in range(len(arr)):
    maxi=-math.inf
    for y in range(len(arr[x])):
        if maxi<arr[y][x]:
            maxi=arr[y][x]
    maxOfcol.append(maxi)
print(maxOfcol) 
```

```python
##Exercise 15: List Product Excluding Duplicates.
a = [1, 3, 5, 6, 3, 5, 6, 1]
l1=[]
product=1
for  item in a:
    if item  not in l1:
        l1.append(item)
for i in l1:
    product*=i
print("Product Excluding Duplication : ",product)    
```

```python
##Exercise 16: Extract elements with Frequency greater than K. 
test_list = [4, 6, 4, 3, 3, 4, 3, 4,4,4,4,4,4,4,4,4668764, 567,7654,7654,6543,76,5,65,65,5,5,6,656,5,65,65,6,53, 8]
k=int(input("enter a target frequency"))
extracted_list=[]
for item in range(len(test_list)):
    count=0
    for i in range(len(test_list)):
        if test_list[item]==test_list[i]:
            count+=1
    if count>k and test_list[item] not in extracted_list:
        extracted_list.append(test_list[item])
print(extracted_list)        
```

```python
##Exercise 18: Break a List into Chunks of Size N. 
n = int(input("Enter your chunk size: "))
test = [1234, 987654, 12345, 75, 76,54, 54,4, 6, 3454,5]

test1 = []

# Step through the list in jumps of size n
for i in range(0, len(test), n):
    temp = test[i:i+n]   # slice from i to i+n
    test1.append(temp)

print(test1)    
```
