```python
##Exercise 1: Perform List Manipulation
my_list = [10, 20, 30, 40, 50]
# 1) Change Element: Change the second element of a list to 200 and print the updated list.
my_list[1]=200
print(my_list)
# 2) Append Element: Add 600 o the end of a list and print the new list.
my_list.append(600)
print(my_list)
# 3) Insert Element: Insert 300 at the third position (index 2) of a list and print the result.
my_list.insert(2,300)
print(my_list)
# 4) Remove Element (by value): Remove 600 from the list and print the list.
my_list.remove(600)
print(my_list)
# 5) Remove Element (by index): Remove the element at index 0 from the list print the list
my_list.pop(0)
print(my_list)
```

```python
##Exercise 2: Sum and average of all numbers in a list
my_list = [10, 20, 30, 40, 50]
total=0
avg=0
for i in range(len(my_list)):
    total+=my_list[i]
avg=total/len(my_list)
print("Total: ",total)
print("Average: ",avg)
```

```python
##Exercise 4: Turn every item of a list into its square
numbers = [1, 2, 3, 4, 5, 6, 7]
rev=[i**2 for i in numbers] # or rev=[numbers[i]**2 for i in range(len(numbers))]
print(rev)
```

```python
##Exercise 5: Find Maximum and Minimum
#Find and print the largest and smallest number in a list [8, 2, 15, 1, 9].
import math
MAX=-math.inf
MIN=math.inf
num=[8,2,15,1,9]
for i in range(len(num)):
    if MAX<num[i]:
        MAX=num[i]
    if MIN>num[i]:
        MIN=num[i]
print("max value: ",MAX)
print("min value: ",MIN)
```

```python
##Exercise 6: Count Occurrences
#Count and print how many times 'Football' appears in list.
sports = ['Cricket', 'Football', 'Hockey', 'Football', 'Tennis']
count=0
if len(sports)>0:
    for i in sports:
        if i=='Football':
            count+=1
print("Occurances of the \"Football\" : ",count)            
        
```
