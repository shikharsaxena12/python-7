```python
# Create a dictionary where the keys are tuples representing coordinates, e.g., {(0,0): "Origin", (1,0): "Right"}.

# Dictionary with tuple keys
coordinates = {
    (0, 0): "Origin",
    (1, 0): "Right",
    (0, 1): "Up",
    (1, 1): "Top-Right",
    (-1, 0): "Left",
    (0, -1): "Down"
}

# Accessing values
print(coordinates[(0, 0)])   # Output: Origin
print(coordinates[(1, 0)])   # Output: Right
print(coordinates[(0, 1)])   # Output: Up
```
