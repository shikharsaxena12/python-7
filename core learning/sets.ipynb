```python
# Take a list from user and print the number of unique elements.
import ast
list_range=int(input("enter a the range of the list which you want:\n"))
l1=[]
for i in range(list_range):
    l1.append(ast.literal_eval(input("enter your {} element".format(i))))
print(set(l1),len(set(l1)))
```
