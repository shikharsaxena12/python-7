```python
# 8. HOTS: Write a program to determine if a number is a palindrome.
# Hint: Reverse the number using string conversion.
n=int(input())
temp=n
rev=0
while temp>0:
    rev=(rev*10)+temp%10
    temp//=10
if rev== n:
    print("Palindrome")
else:
    print("Not a Palindrome")
```

```python
# 14. Print Fibonacci series up to n terms.
# Hint: Use two variables to store previous values.
n=int(input())
a=0
b=1
print(a,b,end=" ")
for i in range(n+1):
    c=a+b
    a=b
    b=c
    print(c,end=" ")    
    
```

```python
# 18. HOTS: Check whether a number is prime.
# Hint: Check divisibility up to sqrt(n).
n=int(input())
count=0
if n<2:
    print("non prime")
else:
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            count+=1
        
if count>0:
    print("not prime")
else:
    print("prime")
```
