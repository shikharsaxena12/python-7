```python
# 18. HOTS: Check whether a number is prime.
# Hint: Check divisibility up to sqrt(n).
n=int(input())
count=0
if n<2:
    print("non prime")
else:
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            count+=1
        
if count>0:
    print("not prime")
else:
    print("prime")
```

```python
# 24. Count frequency of each character.
# Hint: Use dictionary.
string = input("Enter your string: ")

# Create an empty dictionary
freq = {}

# Loop through each character
for char in string:
    if char in freq:
        freq[char] += 1   # increment if already present
    else:
        freq[char] = 1    # add new character with count 1

# Print results
for key, value in freq.items():
    print(f"The frequency of '{key}' is: {value}")
```
