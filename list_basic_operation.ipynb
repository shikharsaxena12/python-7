```python
#create a list  where the range,elements  is also given by the user 
Range_of_list=int(input("enter a range:(integer value only):\n"))
arr=[]
for i in range(Range_of_list):
    n=int(input(f"enter your  index {i}th location element:\n"))
    arr.append(n)##add the element form the last
print("The whole list is here",arr,sep="==>>")
```

```python
### add the element at the specific location
arr.insert(0,"START")
arr.insert(len(arr),"STOP")
print(arr)  
print("new length of the list: ",len(arr))
```

```python
### add the multtiple value iterable in the last of the  list 
arr.extend("shikhar")#- A string is iterable character by character â†’ so "shikhar" becomes ['s','h','i','k','h','a','r'].
arr.extend({"1":2,"2":1})#- Extending with a dictionary iterates over its keys only, not values.
arr.extend({"s","e","t"})
print(arr)
```

```python
### removing the first occurance o the element 
arr.remove("s")
print(arr)
```

```python
### removing the element from the last of the list 
arr.pop()
print(arr)
```

```python
### removing the element for the specific index  location
arr.pop(4)
print(arr)
```

```python
### removing the element for an specific index location  uisng the del keyword
del arr[0]
print(arr)
```

```python
### clear the all element  in the list and return the empty list 
arr.clear()
print(arr)
```

```python
### delete the whole list object 
del arr
print(arr)
```
